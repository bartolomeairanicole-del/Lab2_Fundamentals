```python
 #Setup
correct_password = "PASSWORD123"  # Replace with your required password
attempts = 0
input_matches = False

# Logic: Indefinite iteration
while not input_matches:
    user_input = input("Enter password to unlock system: ")
    attempts += 1
    
    if user_input == correct_password:
        print("System Unlocked.")
        input_matches = True
    else:
        print("Incorrect password. Please try again.")

print(f"Attempts before success: {attempts}")
```
